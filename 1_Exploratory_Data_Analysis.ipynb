# Exploratory Data Analysis Introduction


### Importing Required Libraries:

```python
import ast
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset
import seaborn as sns
```

### Load the required Dataset:

```python
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df.head(5)
```

### Data Cleaning:

```python
# changing the datatype of job_posted_date column from str to timestamp.

df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])

# changing the datatype of job skills column from string to a list.

df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

### Filter for India Data Analyst roles only:

```python
df_IND_DA = df[(df["job_country"] == "India") & (df["job_title_short"] == "Data Analyst")]
df_IND_DA.reset_index(inplace=True)
df_IND_DA
```

```python
df_job_location = df_IND_DA["job_location"].value_counts().head(10).to_frame()
df_job_location
```

```python
df_job_location = df_IND_DA["job_location"].value_counts().head(10).to_frame()

sns.set_theme(style="ticks")
ax = sns.barplot(data=df_job_location, x="count", y="job_location", hue="count", palette="dark:b_r", legend=False)

for i in range(0, 10):
    ax.bar_label(ax.containers[i], fmt="%d")

sns.despine()
plt.title("Number of jobs per job locattion")
plt.xlabel("Jobs Count")
plt.ylabel("Job Locations")
plt.show()
```

```python
df_job_work_from_home = df_IND_DA["job_work_from_home"].value_counts()
df_job_no_degree_mention = df_IND_DA["job_no_degree_mention"].value_counts()
df_job_health_insurance = df_IND_DA["job_health_insurance"].value_counts()



fig, ax = plt.subplots(1, 3)
fig.set_size_inches((12,5))

ax[0].pie(df_job_work_from_home, labels=['False', 'True'], autopct='%1.1f%%', startangle=90)
ax[0].set_title('Work from Home offered')
ax[1].pie(df_job_no_degree_mention, labels=['False', 'True'], autopct='%1.1f%%', startangle=90)
ax[1].set_title('Degree Requirement')
ax[2].pie(df_job_health_insurance, labels=['False'])
ax[2].set_title('Health Insurance offered')

plt.show()
```

```python
df_company_name = df_IND_DA["company_name"].value_counts().head(15).to_frame()
df_company_name
```

```python
df_company_name = df_IND_DA["company_name"].value_counts().head(15).to_frame()

sns.set_theme(style="ticks")
ax = sns.barplot(data=df_company_name, x="count", y="company_name", hue="count", palette="dark:b_r", legend=False)

for i in range(0, 14):
    ax.bar_label(ax.containers[i], fmt="%d")

sns.despine()
plt.title("Number of jobs per Company")
plt.xlabel("Jobs Count")
plt.ylabel("")
plt.show()
```
