# What are the most demanded skills for the most popular Data Science roles

### Importing Libraries, Loading datasets, Data cleanup:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset
import seaborn as sns
import ast


dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)


```

### Filter for India Data Science roles only:

```python
df_India = df[df['job_country']=="India"]

df_India
```

```python
df_India_exploded = df_India.explode('job_skills')
df_India_exploded


```

### Finding the top 3 most popular Data Science roles in India:

```python
df_job_count = df_India.groupby('job_title_short').size().sort_values(ascending=False).to_frame(name="job_count")
df_job_count.head(3)
```

```python
df_skills_count = df_India_exploded.groupby(["job_title_short", "job_skills"]).size()
df_skills_count = df_skills_count.reset_index(name="skills_count")
df_skills_count.sort_values(by="skills_count", ascending=False, inplace=True)
df_skills_count
```

### Skills demand for each of the Data Science roles:

```python
df_skills_count_DE = df_skills_count[df_skills_count["job_title_short"]=="Data Engineer"]
df_skills_count_DE.set_index('job_skills', inplace=True)
df_skills_count_DE.head(5)
```

```python
df_skills_count_DS = df_skills_count[df_skills_count["job_title_short"]=="Data Scientist"]
df_skills_count_DS.set_index('job_skills', inplace=True)
df_skills_count_DS.head(5)
```

```python
df_skills_count_DA = df_skills_count[df_skills_count["job_title_short"]=="Data Analyst"]
df_skills_count_DA.set_index('job_skills', inplace=True)
df_skills_count_DA.head()
```

```python
fig, ax = plt.subplots(3, 1)

sns.despine()
df_skills_count_DE.head().plot(kind="barh", ax=ax[0], xlabel="Jobs Count", ylabel="", title="Data Engineer", legend=False)
ax[0].invert_yaxis()

df_skills_count_DS.head().plot(kind="barh", ax=ax[1], xlabel="Jobs Count", ylabel="", title="Data Scientist", legend=False)
ax[1].invert_yaxis()

df_skills_count_DA.head().plot(kind="barh", ax=ax[2], xlabel="Jobs Count", ylabel="", title="Data Analyst", legend=False)
ax[2].invert_yaxis()

fig.tight_layout()
plt.show()




```
