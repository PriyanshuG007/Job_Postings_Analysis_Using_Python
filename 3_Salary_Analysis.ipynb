# How well do companies pay for Data Science roles

### Importing Libraries, Loading datasets, Data cleanup:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset
import seaborn as sns
import ast


dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_IND = df[(df["job_country"] == "India")].dropna(subset=['salary_year_avg'])
df_IND
```

### List of most popular Data Science roles in India:


```python
df_India = df[df['job_country']=="India"]
top_job_roles = df_India['job_title_short'].value_counts().head(6).index.to_list()
top_job_roles
```

```python
df_plot = df_IND[df_IND["job_title_short"].isin(top_job_roles)]
df_plot
```

```python
job_order = df_plot.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index
job_order
```

```python
sns.boxplot(data=df_plot, x='salary_year_avg', y='job_title_short', order=job_order)
sns.set_theme(style='ticks')


# this is all the same
plt.title('Salary Distributions of Data Jobs in the India')
plt.xlabel('Yearly Salary (USD)')
plt.ylabel('')
plt.xlim(0, 300000) 
ticks_x = plt.FuncFormatter(lambda y, pos: f'${int(y/1000)}K')
plt.gca().xaxis.set_major_formatter(ticks_x)
plt.show()
```
